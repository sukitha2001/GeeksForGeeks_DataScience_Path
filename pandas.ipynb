```python
# Merging DataFrames Using One Key

data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1)

df1 = pd.DataFrame(data2)


print(df, "\n\n", df1)
```

```python
res = pd.merge(df,df1, on='key')
res
```

```python
# Joining DataFrames with Different Index Levels (Multi-Index

data1 = {'Name':['Jai', 'Princi', 'Gaurav'],
        'Age':[27, 24, 22]}

data2 = {'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kanpur'],
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1, index=pd.Index(['K0', 'K1', 'K2'], name='key'))

index = pd.MultiIndex.from_tuples([('K0', 'Y0'), ('K1', 'Y1'),
                                   ('K2', 'Y2'), ('K2', 'Y3')],
                                   names=['key', 'Y'])

df1 = pd.DataFrame(data2, index= index)


print(df, "\n\n", df1)
```

```python
result = df.join(df1, how='inner')
result
```

## Create a Pivot Table in Python using Pandas

```python
df = pd.DataFrame({'Product': ['Carrots', 'Broccoli', 'Banana', 'Banana',
                               'Beans', 'Orange', 'Broccoli', 'Banana'],
                   'Category': ['Vegetable', 'Vegetable', 'Fruit', 'Fruit',
                                'Vegetable', 'Fruit', 'Vegetable', 'Fruit'],
                   'Quantity': [8, 5, 3, 4, 5, 9, 11, 8],
                   'Amount': [270, 239, 617, 384, 626, 610, 62, 90]})
df
```

### Get the Total Sales of Each Product

```python
pivot_table = df.pivot_table(index=['Product'],
                             values=['Quantity'],
                             aggfunc='sum')
print(pivot_table)
```

### Get the Total Sales of Each Category

```python
pivot_table = pd.pivot_table(df,index=['Category'],
                             values=['Amount'],
                             aggfunc='sum')
print(pivot_table)
```

```python
# Get Total Sales by Category and Product Both

pivot_table = pd.pivot_table(df, index=['Category', 'Product'],
                             values=['Amount'],
                             aggfunc='sum')
print(pivot_table)
```

```python
#Get the Mean, Median, Minimum Sale by Category

pivot_table = pd.pivot_table(df, index=['Category'],
                             values=['Amount'],
                             aggfunc=['mean', 'median', 'min'])
print(pivot_table)
```

```python
#Get the Mean, Median, Minimum Sale by Product

pivot_table = pd.pivot_table(df, index=['Product'],
                             values=['Amount'],
                             aggfunc=['mean', 'median', 'min'])
print(pivot_table)
```

## Binary operations on Pandas DataFrame and Series

```python
s1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
s2 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])

# Adding the two Series

print(s1)
print('-----------')
print(s2)
print('-----------')
result = s1 + s2
print(result)
```

```python
# Comparison Operations on Series

s1 = pd.Series([10,20,30])
s2 = pd.Series([10,25,30])

result = s1 == s2
print(result)
```
