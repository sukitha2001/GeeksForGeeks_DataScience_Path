```python
# Creating a DataFrame from a List of Dictionaries 

dict = {'names': ['Sukitha', 'Prabhavi', 'Vidura', 'Mokshana'],
        'ages' : [23, 22, 21, 20],
        'city' : ['Colombo', 'Kandy', 'Galle', 'Jaffna']}
df = pd.DataFrame(dict)
print(df)
```

```python
#Accessing and Modifying the Index

print(df.index)
```

```python
#Setting a Custom Index
# Set 'Name' column as the index

df_with_index = df.set_index('names')
print(df_with_index)
```

```python
#Resetting the Index

df_reset = df_with_index.reset_index()
print(df_reset)
```

```python
#Indexing with loc

data = {'Name': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'],
        'Age': [25, 30, 22, 35, 28],
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'Salary': [50000, 55000, 40000, 70000, 48000]}

df = pd.DataFrame(data)
```

```python
row = df.loc[1]
print(row)
```

```python
#Accessing Columns From DataFrame

age_column = df['Age']
print(age_column)
```

```python
#Accessing Rows by Index

second_row = df.iloc[1]
print(second_row)
```

```python
# Accessing Multiple Rows or Columns

subset = df.loc[0:2, ['Name', 'Age']]
print(subset)
```

```python
# Accessing Rows Based on Conditions

filtered_data = df[df['Age']>25]
print(filtered_data)
```

```python
#  Accessing Specific Cells with at and iat

salary_at_index_2 =df.at[2, 'Salary']
print(salary_at_index_2)
```
